```python
# @title 1. åˆå§‹åŒ–ç¯å¢ƒ & æ³¨å…¥é˜²å´©æºƒç–«è‹—
# ==============================================================================
# åŠŸèƒ½ï¼šå®‰è£…ä¾èµ–ï¼Œä¿®å¤ T4 æ˜¾å¡è¯†åˆ«é—®é¢˜ï¼Œå¹¶ä¿®æ”¹ BabelDOC æºç ä»¥é˜²æ­¢è¶…æ—¶å´©æºƒã€‚
# ==============================================================================
import os
import re
import time

print("ğŸ—ï¸ [1/3] æ­£åœ¨æ­å»º WSJ ç¿»è¯‘æµæ°´çº¿...")

# 1. ç³»ç»Ÿçº§ä¾èµ–å®‰è£… (è®© Colab çœ‹è§æ˜¾å¡)
os.system("sudo apt-get install -y pciutils lshw zstd")

# 2. Python ä¾èµ–ä¿®å¤ (å¼ºåˆ¶é™çº§ Numpy ä»¥å…¼å®¹ PaddleOCR)
print("ğŸ”§ æ­£åœ¨é”å®šä¾èµ–åº“ç‰ˆæœ¬...")
os.system("pip install paddlepaddle-gpu")
os.system("pip install 'numpy<2.0' --force-reinstall")

# 3. å®‰è£… BabelDOC
if not os.path.exists('BabelDOC'):
    print("ğŸ“¦ å…‹éš† BabelDOC ä»“åº“...")
    os.system("git clone https://github.com/funstory-ai/BabelDOC.git")
    os.system("pip install -r BabelDOC/requirements.txt")
    os.system("pip install -e BabelDOC")

# 4. å®‰è£… Ollama æœåŠ¡
print("â¬‡ï¸ å®‰è£… Ollama AI å¼•æ“...")
os.system("curl -fsSL https://ollama.com/install.sh | sh")

# 5. ã€æ ¸å¿ƒã€‘æ³¨å…¥â€œæºç çº§é˜²å´©æºƒè¡¥ä¸â€
# è¯´æ˜ï¼šBabelDOC åŸç”Ÿä»£ç æ²¡æœ‰è¶…æ—¶é‡è¯•æœºåˆ¶ï¼Œé‡åˆ°éš¾å•ƒçš„æ®µè½ä¼šå¡æ­»ã€‚
# æˆ‘ä»¬ç›´æ¥ç”¨æ­£åˆ™æ›¿æ¢ translator.py ä¸­çš„æ ¸å¿ƒå‡½æ•°ï¼Œæ¤å…¥â€œè¶…æ—¶ç†”æ–­+é‡è¯•â€é€»è¾‘ã€‚
print("ğŸ’‰ æ­£åœ¨æ‰§è¡Œæºç æ‰‹æœ¯ (Timeout Patch)...")
target_file = '/content/BabelDOC/babeldoc/translator/translator.py'

if os.path.exists(target_file):
    with open(target_file, 'r') as f:
        code = f.read()

    # å®šä¹‰â€œæ— æ•Œç‰ˆâ€ç¿»è¯‘å‡½æ•°
    robust_func = """def do_llm_translate(self, text, system_prompt=None):
        import time
        max_retries = 3
        for attempt in range(max_retries):
            try:
                messages = []
                if system_prompt:
                    messages.append({"role": "system", "content": system_prompt})
                messages.append({"role": "user", "content": text})

                # [è¡¥ä¸] å¼ºåˆ¶ 120ç§’ è¶…æ—¶ï¼Œé˜²æ­¢æ­»é”
                response = self.client.chat.completions.create(
                    model=self.model,
                    messages=messages,
                    temperature=0.1,  # ç»å¯¹ç†æ€§æ¨¡å¼
                    timeout=120       # ç†”æ–­æ—¶é—´
                )
                return response.choices[0].message.content
            except Exception as e:
                print(f"\\nğŸ›¡ï¸ [æŠ¤ç›¾æ¿€æ´»] ç¿»è¯‘å°è¯• {attempt+1}/{max_retries} å¤±è´¥ï¼Œè‡ªåŠ¨é‡è¯•...")
                time.sleep(2)

        print("âŒ [è·³è¿‡] è¯¥æ®µè½è¿‡äºå¤æ‚æˆ–å«ä¹±ç ï¼Œå·²è·³è¿‡ã€‚")
        return "[åŸæ–‡ç¿»è¯‘å¤±è´¥ï¼Œå·²è·³è¿‡]"
"""
    # æ­£åˆ™æ›¿æ¢
    pattern = r"def do_llm_translate\(self, text, system_prompt=None\):[\s\S]*?return response\.choices\[0\]\.message\.content"
    new_code = re.sub(pattern, robust_func, code)

    with open(target_file, 'w') as f:
        f.write(new_code)
    print("âœ… è¡¥ä¸æ³¨å…¥å®Œæˆï¼ç³»ç»Ÿå·²å‡çº§ä¸ºâ€œéŸ§æ€§æ¨¡å¼â€ã€‚")
else:
    print("âŒ è­¦å‘Šï¼šæœªæ‰¾åˆ°æºç æ–‡ä»¶ï¼Œè¡¥ä¸å¯èƒ½æœªç”Ÿæ•ˆã€‚")

print("\nğŸ‰ ç¯å¢ƒå‡†å¤‡å°±ç»ªï¼è¯·ã€é‡å¯ä¼šè¯ (Restart Session)ã€‘åè¿è¡Œç¬¬ 2 æ­¥ã€‚")
```

```python
# @title 2. å¯åŠ¨æœåŠ¡ & é…ç½® Google ç¿»è¯‘æ¨¡å‹
# ==============================================================================
# åŠŸèƒ½ï¼šå¯åŠ¨åå°æœåŠ¡ï¼Œå¹¶åˆ›å»ºä¸€ä¸ªä¸“ä¸º WSJ ä¼˜åŒ–çš„ TranslateGemma-32k æ¨¡å‹ã€‚
# ==============================================================================
import subprocess
import time
import requests

print("ğŸš€ [2/3] æ­£åœ¨é¢„çƒ­ AI å¼•æ“...")

# 1. å¯åŠ¨åå°æœåŠ¡
subprocess.run(["pkill", "ollama"]) # æ¸…ç†æ—§è¿›ç¨‹
process = subprocess.Popen(["ollama", "serve"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

# ç­‰å¾…æœåŠ¡å“åº”
print("â³ ç­‰å¾… Ollama æœåŠ¡ä¸Šçº¿...", end="")
for i in range(20):
    try:
        if requests.get("http://localhost:11434").status_code == 200:
            print("\nâœ… æœåŠ¡å·²è¿é€šï¼")
            break
    except:
        time.sleep(2)
        print(".", end="")

# 2. æ¨¡å‹ç»„è£…è½¦é—´
print("ğŸ’ æ­£åœ¨ç»„è£… Google TranslateGemma (4B)...")
subprocess.run(["ollama", "pull", "translategemma:4b"])

print("âš™ï¸ æ­£åœ¨åº”ç”¨ WSJ ä¸“ç”¨å‚æ•° (32k Context)...")
# åˆ›å»ºä¸€ä¸ªåä¸º "wsj-translator" çš„å®šåˆ¶æ¨¡å‹
with open("Modelfile_wsj", "w") as f:
    f.write("FROM translategemma:4b\n")
    f.write("PARAMETER num_ctx 32768\n")      # 32k è¶…å¤§æ˜¾å­˜çª—å£
    f.write("PARAMETER repeat_penalty 1.2\n") # é˜²æ­¢å¤è¯»æœºæ­»å¾ªç¯
    f.write("PARAMETER temperature 0.1")      # é™ä½åˆ›é€ æ€§ï¼Œæé«˜å‡†ç¡®åº¦
subprocess.run(["ollama", "create", "wsj-translator", "-f", "Modelfile_wsj"])

print("\nğŸ‰ ä¸“ç”¨ç¿»è¯‘å¼•æ“ 'wsj-translator' å‡†å¤‡å°±ç»ªï¼")
```

```python
# @title 3. (åŒä¿é™©ç‰ˆ) è‡ªåŠ¨æ¥åŠ›ç¿»è¯‘ + Google Drive å®æ—¶å­˜æ¡£
# ==============================================================================
# åŠŸèƒ½ï¼šè‡ªåŠ¨æŒ‚è½½ Google Driveã€‚æ¯ç¿»è¯‘å®Œ 5 é¡µï¼Œå…ˆå­˜å…¥ç½‘ç›˜ï¼ˆé˜²æ–­å¼€ï¼‰ï¼Œå†å°è¯•ä¸‹è½½åˆ°æœ¬åœ°ã€‚
# ==============================================================================
import os
import subprocess
import time
import shutil
from google.colab import files
from google.colab import drive

# === ç”¨æˆ·é…ç½® ===
BATCH_SIZE = 5    # é»„é‡‘å®šå¾‹ï¼šæ¯ 5 é¡µæ¸…ç†ä¸€æ¬¡
TOTAL_PAGES = 20  # æ€»é¡µæ•°

# 1. æŒ‚è½½ Google Drive (æ ¸å¿ƒä¿é™©æªæ–½)
print("â˜ï¸ æ­£åœ¨è¿æ¥ Google Drive (äº‘ç«¯ä¿é™©ç®±)...")
try:
    drive.mount('/content/drive')
    # åˆ›å»ºå­˜æ¡£æ–‡ä»¶å¤¹
    drive_folder = "/content/drive/My Drive/WSJ_Translation_Result"
    if not os.path.exists(drive_folder):
        os.makedirs(drive_folder)
    print(f"âœ… ç½‘ç›˜æŒ‚è½½æˆåŠŸï¼æ–‡ä»¶å°†è‡ªåŠ¨ä¿å­˜åˆ°: {drive_folder}")
except Exception as e:
    print(f"âš ï¸ ç½‘ç›˜æŒ‚è½½å¤±è´¥ (å¯èƒ½æ˜¯æƒé™é—®é¢˜)ï¼Œå°†åªä½¿ç”¨æœ¬åœ°ä¸‹è½½ã€‚é”™è¯¯: {e}")
    drive_folder = None

# 2. ä¸Šä¼ æ–‡ä»¶
print("\nğŸ“‚ è¯·ä¸Šä¼  WSJ.pdf...")
uploaded = files.upload()
if not uploaded:
    raise SystemExit("âŒ æœªä¸Šä¼ æ–‡ä»¶")
filename = list(uploaded.keys())[0]
base_name = os.path.splitext(filename)[0]

# å·¥å…·å‡½æ•°ï¼šé‡å¯æœåŠ¡
def reset_engine():
    print("ğŸ§¹ [å†…å­˜å›æ”¶] æ­£åœ¨é‡å¯æœåŠ¡ä»¥é‡Šæ”¾æ˜¾å­˜...")
    subprocess.run(["pkill", "ollama"])
    time.sleep(2)
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    time.sleep(5)

print(f"\nâš”ï¸ å¯åŠ¨åŒä¿é™©æ¥åŠ›ï¼šç›®æ ‡ {TOTAL_PAGES} é¡µã€‚")

for start in range(1, TOTAL_PAGES + 1, BATCH_SIZE):
    end = min(start + BATCH_SIZE - 1, TOTAL_PAGES)
    page_range = f"{start}-{end}"

    # 1. æ¸…æ´—å¼•æ“
    reset_engine()

    print(f"\nğŸš€ [ç¬¬ {page_range} é¡µ] æ­£åœ¨ç¿»è¯‘ä¸­...")

    # 2. æ‰§è¡Œç¿»è¯‘
    cmd = f"""
    babeldoc \
      --files "{filename}" \
      --pages "{page_range}" \
      --openai \
      --openai-model "wsj-translator" \
      --openai-base-url "http://localhost:11434/v1" \
      --openai-api-key "ollama" \
      --lang-out zh \
      --no-auto-extract-glossary \
      --custom-system-prompt "Translate into Simplified Chinese."
    """
    exit_code = os.system(cmd)

    # 3. å­˜æ¡£ä¸ä¸‹è½½
    if exit_code == 0:
        # åŸå§‹æ–‡ä»¶å
        original_dual = f"{base_name}.zh.dual.pdf"
        original_mono = f"{base_name}.zh.mono.pdf"

        # æ–°æ–‡ä»¶å
        new_dual = f"WSJ_Part_{page_range}_Dual.pdf"
        new_mono = f"WSJ_Part_{page_range}_Mono.pdf"

        # (A) é‡å‘½å
        if os.path.exists(original_dual):
            shutil.move(original_dual, new_dual)
        if os.path.exists(original_mono):
            shutil.move(original_mono, new_mono)

        # (B) å­˜å…¥ Google Drive (æœ€ç¨³å¦¥çš„ä¸€æ­¥)
        if drive_folder:
            print(f"â˜ï¸ æ­£åœ¨åŒæ­¥åˆ° Google Drive...")
            if os.path.exists(new_dual):
                shutil.copy(new_dual, os.path.join(drive_folder, new_dual))
            if os.path.exists(new_mono):
                shutil.copy(new_mono, os.path.join(drive_folder, new_mono))
            print("âœ… å·²å®‰å…¨å­˜å…¥ç½‘ç›˜ï¼")

        # (C) å°è¯•ä¸‹è½½åˆ°æœ¬åœ° (å¦‚æœæµè§ˆå™¨è¿˜è¿ç€)
        print("â¬‡ï¸ å°è¯•å‘é€åˆ°æœ¬åœ°...")
        try:
            files.download(new_dual)
            files.download(new_mono)
        except Exception:
            print("âš ï¸ æµè§ˆå™¨è‡ªåŠ¨ä¸‹è½½å¤±è´¥ (å¯èƒ½çª—å£å·²ä¼‘çœ )ï¼Œè¯·å» Drive æŸ¥çœ‹ã€‚")

    else:
        print(f"âš ï¸ ç¬¬ {page_range} é¡µç¿»è¯‘å¼‚å¸¸ï¼Œè·³è¿‡ã€‚")

print(f"\nğŸ å…¨éƒ¨å®Œæˆï¼å³ä½¿è¿æ¥æ–­å¼€ï¼Œæ‚¨çš„æ–‡ä»¶ä¹Ÿåœ¨ Google Drive çš„ 'WSJ_Translation_Result' æ–‡ä»¶å¤¹é‡Œã€‚")
```
